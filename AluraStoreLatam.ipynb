# Alura Store Latam

Este notebook desarrolla el challenge de Alura para comparar el desempeno de las 4 tiendas y recomendar cual deberia vender el Sr. Juan.

El analisis cubre:
- ingreso total por tienda,
- ventas por categoria,
- calificacion promedio,
- productos mas y menos vendidos,
- costo de envio promedio.


### Importacion de datos


```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

## Preparacion de los datos

Se crea una estructura comun para trabajar las 4 tiendas con las mismas columnas y obtener un resumen general antes de entrar a cada analisis.


```python
import unicodedata

plt.style.use("seaborn-v0_8")

def normalizar_texto(texto):
    return unicodedata.normalize("NFKD", texto).encode("ascii", "ignore").decode("utf-8")

def normalizar_columnas(df):
    df = df.copy()
    df.columns = [normalizar_texto(columna) for columna in df.columns]
    return df

tiendas = {
    "Tienda 1": normalizar_columnas(tienda),
    "Tienda 2": normalizar_columnas(tienda2),
    "Tienda 3": normalizar_columnas(tienda3),
    "Tienda 4": normalizar_columnas(tienda4),
}

resumen_tiendas = pd.DataFrame(
    {
        "Ingreso total": {nombre: df["Precio"].sum() for nombre, df in tiendas.items()},
        "Ticket promedio": {nombre: df["Precio"].mean() for nombre, df in tiendas.items()},
        "Calificacion promedio": {nombre: df["Calificacion"].mean() for nombre, df in tiendas.items()},
        "Envio promedio": {nombre: df["Costo de envio"].mean() for nombre, df in tiendas.items()},
        "Cantidad de ventas": {nombre: len(df) for nombre, df in tiendas.items()},
    }
).round(2).sort_values("Ingreso total", ascending=False)

display(resumen_tiendas)
```

## 1. Analisis de facturacion

El primer paso es comparar el ingreso total generado por cada tienda a partir de la columna `Precio`.


```python
ingresos_totales = resumen_tiendas["Ingreso total"].sort_values(ascending=False)
display(ingresos_totales)

fig, ax = plt.subplots(figsize=(8, 4))
colors = ["#1d3557", "#457b9d", "#a8dadc", "#e76f51"]
bars = ax.bar(ingresos_totales.index, ingresos_totales.values, color=colors)
ax.set_title("Ingreso total por tienda")
ax.set_ylabel("Ingreso total")
ax.ticklabel_format(style="plain", axis="y")
ax.bar_label(bars, labels=[f"${valor:,.0f}" for valor in ingresos_totales.values], padding=3, fontsize=9)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

**Hallazgo clave:** La Tienda 1 lidera en ingresos con aproximadamente `$1,150,880,400`, mientras que la Tienda 4 queda ultima con `$1,038,375,700`. La diferencia supera los `$112 millones`, lo que ya la deja como la candidata mas debil en terminos de facturacion.


## 2. Ventas por categoria

Ahora se revisa el volumen de ventas por categoria para identificar si alguna tienda destaca o se rezaga en su mezcla de productos.


```python
ventas_categoria = pd.DataFrame(
    {
        nombre: df["Categoria del Producto"].value_counts()
        for nombre, df in tiendas.items()
    }
).fillna(0).astype(int)

display(ventas_categoria.sort_values("Tienda 4", ascending=False))

categoria_t4 = ventas_categoria["Tienda 4"].sort_values(ascending=False)
pie_t4 = pd.concat(
    [categoria_t4.head(5), pd.Series({"Otras categorias": categoria_t4.iloc[5:].sum()})]
)

fig, ax = plt.subplots(figsize=(7, 7))
ax.pie(
    pie_t4.values,
    labels=pie_t4.index,
    autopct="%1.1f%%",
    startangle=90,
    colors=["#264653", "#2a9d8f", "#8ab17d", "#e9c46a", "#f4a261", "#e76f51"],
)
ax.set_title("Distribucion de ventas por categoria en la Tienda 4")
plt.tight_layout()
plt.show()
```

**Hallazgo clave:** En las 4 tiendas se repite el mismo patron: `Muebles` y `Electronicos` dominan el volumen de ventas. La Tienda 4 no presenta una categoria diferencial que compense su menor facturacion total, por lo que su mezcla comercial luce menos competitiva que la del resto.


## 3. Calificacion promedio de la tienda

La satisfaccion del cliente se aproxima usando el promedio de la columna `Calificacion`.


```python
calificaciones = resumen_tiendas["Calificacion promedio"].sort_values(ascending=False)
display(calificaciones)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(calificaciones.index, calificaciones.values, marker="o", linewidth=2.5, color="#2a9d8f")
ax.set_title("Calificacion promedio por tienda")
ax.set_ylabel("Calificacion promedio")
ax.set_ylim(calificaciones.min() - 0.02, calificaciones.max() + 0.02)
ax.grid(axis="y", linestyle="--", alpha=0.4)
for i, valor in enumerate(calificaciones.values):
    ax.text(i, valor + 0.003, f"{valor:.2f}", ha="center")
plt.tight_layout()
plt.show()
```

**Hallazgo clave:** La mejor valorada es la Tienda 3 (`4.05`), seguida por la Tienda 2 (`4.04`). La Tienda 1 es la peor valorada (`3.98`) y la Tienda 4 queda ligeramente por debajo del bloque fuerte con `4.00`, lo que refuerza que no destaca tampoco en experiencia del cliente.


## 4. Productos mas y menos vendidos

Este analisis sirve para detectar cuales productos concentran mas demanda y cuales tienen una rotacion baja en cada tienda.


```python
def extremos_ventas(df, limite=3):
    conteo = df["Producto"].value_counts()
    mas_vendidos = conteo[conteo == conteo.max()].sort_index().head(limite)
    menos_vendidos = conteo[conteo == conteo.min()].sort_index().head(limite)
    return mas_vendidos, menos_vendidos

for nombre, df in tiendas.items():
    mas_vendidos, menos_vendidos = extremos_ventas(df)
    print(nombre)
    print("  Mas vendidos:")
    for producto, cantidad in mas_vendidos.items():
        print(f"    - {producto}: {cantidad} ventas")
    print("  Menos vendidos:")
    for producto, cantidad in menos_vendidos.items():
        print(f"    - {producto}: {cantidad} ventas")
    print()
```

**Hallazgo clave:** La Tienda 4 tiene como producto mas vendido la `Cama box`, mientras que su producto menos vendido es la `Guitarra electrica`. Esta metrica ayuda a entender la rotacion, pero por si sola no compensa su bajo ingreso total ni su menor ticket promedio.


## 5. Envio promedio por tienda

Finalmente, se compara el costo de envio promedio para ver si existe alguna ventaja operativa relevante entre las tiendas.


```python
envio_vs_calificacion = resumen_tiendas[
    ["Envio promedio", "Calificacion promedio", "Ingreso total"]
]
display(envio_vs_calificacion.sort_values("Ingreso total", ascending=False))

fig, ax = plt.subplots(figsize=(8, 5))
sizes = envio_vs_calificacion["Ingreso total"] / 8_000_000
ax.scatter(
    envio_vs_calificacion["Envio promedio"],
    envio_vs_calificacion["Calificacion promedio"],
    s=sizes,
    c=["#1d3557", "#457b9d", "#2a9d8f", "#e76f51"],
    alpha=0.8,
)

for nombre, fila in envio_vs_calificacion.iterrows():
    ax.annotate(
        nombre,
        (fila["Envio promedio"], fila["Calificacion promedio"]),
        textcoords="offset points",
        xytext=(6, 6),
    )

ax.set_title("Relacion entre envio promedio, calificacion e ingreso total")
ax.set_xlabel("Envio promedio")
ax.set_ylabel("Calificacion promedio")
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

**Hallazgo clave:** La Tienda 4 tiene el envio promedio mas bajo (`$23,459.46`), pero esa ventaja no se convierte en mejores ingresos ni en una posicion claramente superior frente a las otras tiendas. La Tienda 1, por ejemplo, carga con el envio mas alto, pero aun asi conserva el mejor nivel de facturacion.


## Informe final


Despues de comparar ingresos, categorias, calificaciones, productos y costos de envio, la recomendacion es vender la **Tienda 4**.

La razon principal es que presenta el **menor ingreso total** del grupo (`$1,038,375,700`) y tambien el **ticket promedio mas bajo** (`$440,362.89`). Aunque su costo de envio promedio es el mas conveniente, esa ventaja no alcanza para compensar su debilidad financiera.

Ademas, la Tienda 4 tampoco sobresale en satisfaccion del cliente: su **calificacion promedio** (`4.00`) queda por debajo de la Tienda 2 y la Tienda 3. En ventas por categoria mantiene una distribucion similar a la del resto, sin una fortaleza clara que justifique conservarla frente a alternativas con mejor equilibrio entre ingresos y experiencia del cliente.

La Tienda 1 tiene retos importantes en calificacion y envio, pero sigue siendo la de mayor facturacion. Por eso, considerando el desempeno general del negocio, la opcion mas coherente para vender es la **Tienda 4**.
